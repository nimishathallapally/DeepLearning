```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

```python
imgSize = 224
patchSize = 16
numClasses = 2
batchSize = 32
autoTune = tf.data.AUTOTUNE
```

### Optimizing Data Pipelines with `tf.data.AUTOTUNE`

When you're building a data pipeline with `tf.data`, you often use operations like:

- `dataset.map(...)`
- `dataset.prefetch(...)`
- `dataset.cache()`

Some of these operations take arguments such as `num_parallel_calls` or `buffer_size`.  
If you pass `tf.data.AUTOTUNE`, TensorFlow will **automatically determine the optimal value** based on your system’s CPU/GPU and workload.

```python
# Load image paths and Labels form DataFrames

def pathLabelDs(df):
    paths = df['Image Path'].values
    labels = df['Label'].values
    return paths, labels

trainPaths, trainLabels = pathLabelDs(train_df)
valPaths, valLabels = pathLabelDs(val_df)
testPaths, testLabels = pathLabelDs(test_df)
```

```python
# Image Loading Function
def preprocessImage(filePath, label):
    img = tf.io.read_file(filePath)
    img = tf.image.decode_png(img, channels = 3)    # Decode the image as a PNG file with 3 color channels
    img = tf.image.resize(img, [imgSize, imgSize])
    img = tf.cast(img, tf.float32) / 255.0         # Normalize pixel values to the range [0, 1] (float32)
    return img, label
```

```python
def makeDataset(paths, labels):
    ds = tf.data.Dataset.from_tensor_slices((paths, labels))   # TensorFlow dataset - pairing each file path with its label
    ds = ds.map(preprocessImage, num_parallel_calls = autoTune)     # Apply preprocessing to each image in parallel using Threads
    return ds.batch(batchSize).prefetch(autoTune) #  groups the dataset into batches and Batch the data and prefetch for performance


train_ds = makeDataset(trainPaths, trainLabels)
val_ds = makeDataset(valPaths, valLabels)
test_ds = makeDataset(testPaths, testLabels)
```

```python
# Patch creation layer
class Patches(layers.Layer):
    def __init__(self, patchSize):
        super().__init__()
        self.patchSize = patchSize

    def call(self, images):
        batchSize = tf.shape(images)[0]
        patches = tf.image.extract_patches(
            images = images,
            sizes = [1, self.patchSize, self.patchSize, 1],
            strides = [1, self.patchSize, self.patchSize, 1],
            rates = [1, 1, 1, 1],
            padding = "VALID"
        )
        patch_dims = patches.shape[-1]
        patches = tf.reshape(patches, [batchSize, -1, patch_dims])
        return patches
```

```python
# Positional embedding
class PatchEncoder(layers.Layer):
    def __init__(self, num_patches, projection_dim):
        super().__init__()
        self.projection = layers.Dense(projection_dim)
        self.position_embedding = layers.Embedding(input_dim=num_patches, output_dim=projection_dim)

    def call(self, patch):
        positions = tf.range(start=0, limit=tf.shape(patch)[1], delta=1)
        encoded = self.projection(patch) + self.position_embedding(positions)
        return encoded
```

```python
def create_vit_classifier():
    num_patches = (imgSize // patchSize) ** 2
    projection_dim = 64
    transformer_layers = 8
    num_heads = 4
    transformer_units = [projection_dim * 2, projection_dim]

    inputs = layers.Input(shape=(imgSize, imgSize, 3))
    patches = Patches(patchSize)(inputs)
    encoded_patches = PatchEncoder(num_patches, projection_dim)(patches)

    for _ in range(transformer_layers):
        x1 = layers.LayerNormalization(epsilon=1e-6)(encoded_patches)
        attention_output = layers.MultiHeadAttention(num_heads=num_heads, key_dim=projection_dim)(x1, x1)
        x2 = layers.Add()([attention_output, encoded_patches])
        x3 = layers.LayerNormalization(epsilon=1e-6)(x2)
        x3 = layers.Dense(transformer_units[0], activation='gelu')(x3)
        x3 = layers.Dense(transformer_units[1], activation='gelu')(x3)
        encoded_patches = layers.Add()([x3, x2])

    representation = layers.LayerNormalization(epsilon=1e-6)(encoded_patches)
    representation = layers.GlobalAveragePooling1D()(representation)  # ✅ Fix: use GAP instead of Flatten
    representation = layers.Dropout(0.3)(representation)
    features = layers.Dense(128, activation='gelu')(representation)
    logits = layers.Dense(1, activation='sigmoid')(features)

    model = keras.Model(inputs=inputs, outputs=logits)
    return model
```

```python
# Compile and train
vit_model = create_vit_classifier()
vit_model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=1e-4),
    loss='binary_crossentropy',
    metrics=['accuracy', tf.keras.metrics.AUC(name="auc")]
)
```

```python
history = vit_model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=50
)
```

```python
from sklearn.metrics import classification_report

# Predict on test set
y_pred_probs = vit_model.predict(test_ds)
y_preds = (y_pred_probs > 0.5).astype(int)

# True labels
y_true = np.concatenate([y for x, y in test_ds], axis=0)

# Report
print(classification_report(y_true, y_preds))
```
